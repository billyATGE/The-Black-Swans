```python
!pip install pytrends
import pandas as pd                        
from pytrends.request import TrendReq
from functools import reduce
import dateutil
import matplotlib.pyplot as plt
from scipy.stats import linregress
pytrend = TrendReq()
```

```python
pytrend.build_payload(kw_list=['food near me','Health'],timeframe='2014-12-20 2014-12-31',geo='US')
df_CI_2014 = pytrend.interest_over_time().reset_index()
df_CI_2014['Year'] = df_CI_2014['date'].dt.year
df_CI_2014['Month'] = df_CI_2014['date'].dt.month
df_CI_2014['Day'] = df_CI_2014['date'].dt.day
df_CI_2014 = df_CI_2014.drop(['date','isPartial'],axis=1)
```

```python
cDay_df = [df_CI_2014,df_CI_2015,df_CI_2016,df_CI_2017,df_CI_2018,df_CI_2019,df_CI_2020]
df_final = pd.concat(cDay_df, sort=False)
cDay_final = df_final[['food near me','Health','Day']].groupby('Day').sum().reset_index()
```

```python
x_values = cDay_final['Day'].loc[cDay_final['Day'] <= 24]
y_values = cDay_final['food near me'].loc[cDay_final['Day'] <= 24]
(slope, intercept, rvalue, pvalue, stderr) = linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = 'y = ' + str(round(slope,2)) + 'x + ' + str(round(intercept,2))
print(f'The r-value is: {rvalue}')
plt.plot(x_values,y_values)
plt.plot(x_values,regress_values,'r-')
plt.annotate(line_eq,(22,190),fontsize=15,color='red')
plt.title('Pre Christmas Food Near Me')
plt.show()
```
